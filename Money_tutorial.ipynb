```python
import pandas as pd

from abm.boltzman_money import MoneyModel
```

```python
mm = MoneyModel(10, 10, 5)
for i in range(100):
    mm.step()
```

```python
df = (mm
      .datacollector
      .get_agent_vars_dataframe()
      ["Wealth"]
      .unstack("AgentID")
      )
df.plot(figsize=(20, 10))
```

```python
import pandas as pd
df = (mm
      .datacollector
      .get_agent_vars_dataframe()
      ['Pos']
      .apply(lambda x: pd.Series({'x': x[0], 'y': x[1]}))
      )
df['x'].unstack("AgentID").plot()
```
